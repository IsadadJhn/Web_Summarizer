```python
web= fetch_website_contents("https://sport.detik.com/sepakbola/liga-inggris/d-8462194/main-habis-habisan-arsenal-taklukkan-newcastle-united")
web
```

```python
system_prompt="""You are humoris assistant that analyze the content of website,and provides a short, snarky, humorous summary, ignoring text that might be navigation related.
Respond in markdown. Do not wrap the markdown in a code block - respond just with the markdown. respond in bahasa indonesia"""
```

```python
user_prompt="""
Here are the contents of a website.
Provide a short summary of this website.
"""
```

```python
messages = [
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": user_prompt}
]
```

```python
response = ollama.chat("llama3.2",messages=messages, stream =True)
```

```python
def messages_for(website): # ini berisi url website
    return [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt + website }
    ]
```

```python
messages_for(web)
```

```python
def summarize(url):
    website = fetch_website_contents(url)
    response = ollama.chat(
        model="llama3.2",
        messages=messages_for(website)
    )
    return response ["message"] ["content"] #message dan content berperan sebagai function
```

```python
summarize("https://sport.detik.com/sepakbola/liga-inggris/d-8462194/main-habis-habisan-arsenal-taklukkan-newcastle-united")
```

```python
def display_summary(url):
    summary = summarize(url)
    display(Markdown(summary))
    
```

```python
display_summary("https://sport.detik.com/sepakbola/liga-inggris/d-8462194/main-habis-habisan-arsenal-taklukkan-newcastle-united")
```
